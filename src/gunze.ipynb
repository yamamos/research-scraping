```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.options import Options as ChromeOptions
from webdriver_manager.chrome import ChromeDriverManager
import csv
import time
from selenium.webdriver.common.by import By
import pandas as pd

import numpy as np

from datetime import datetime
```

```python
product_dict = {
                # "vEE3417A": "着心地軽やかなカットオフインナー", # men
                # "vKQ5016": "半袖Ｕ首",
                "vEE3315A": "ＵネックＴシャツ",
                "vKQ5014": "半袖丸首",
                # ############################
                "vYV2608W": "【大山 旬さん監修】in.T(インティー)8.5分袖シャツ",
                "vEE3615A": "ＵネックＴシャツ(深めのUネック)(Ｕ首)",
                "vMH0808": "【極】ロングスリーブシャツ（9分袖）",
                # ############################
                # "vKB9258K": "【Fitte COOL】ブラタンクトップ(ラウンドネック)", # women'
                # "vKB9157K": "【Fitte COOL】ブラキャミソール",
                # "vKL7458R": "【さらさら強撚綿】ラン型インナー（パッド付）",
                # # ############################
                # "vKB8059K": "【Fitte Warm】8分袖ブラトップ",
                # "vMH5946": "【極】8分袖インナー",
                # "vKL9959R": "【綿混起毛】8分袖ウォーマー（パッド付）",

                }
```

```python
gender = "男性"

for product_id in product_dict.keys():

    driver = webdriver.Chrome()
    driver.get(f"https://www.gunze.jp/store/g/{product_id}")

    # 口コミのタブを開く
    time.sleep(3)
    
    elem = driver.find_element(By.XPATH,'/html/body/div[1]/div[4]/div/main/div[1]/div[2]/div[2]/ul/li[3]/span[1]')
    elem.click()

    df = pd.DataFrame()
 
    while True:
        time.sleep(3)

        try:
            html = driver.page_source.encode('utf-8')
            soup = BeautifulSoup(html, "html.parser")
            
            reviews = soup.find_all('div', class_='zv-answer zv-comment')
            stars = soup.find_all('div', class_='zv-answer zv-stars')
            date = soup.find_all('span', class_='zv-posted-at')
            product = product_dict[product_id]

            date_list = [d.get_text() for d in date]
            # date_listの要素をdatetime型に変換
            date_list = [datetime.strptime(d.replace('投稿日\xa0', ''), '%Y年%m月%d日') for d in date_list]


            # starsの要素に"zv-star-svg"が何回含まれるかカウントして、それをリストに格納する
            star_list = []
            for s in stars:
                single_rev_stars = s.find_all('svg')
                single_rev_stars_count = 5 - sum(1 for star in single_rev_stars if "zv-star-svg zv-star-svg-grey" in str(star))
                star_list.append(single_rev_stars_count)

            review_list = [t.get_text() for t in reviews]
            # review_listの要素に¥u3000が含まれている場合は、その要素の¥u3000を削除する
            review_list = [t.replace('\u3000', '') for t in review_list]

            column_dict = {"product": product ,
                        "review": review_list,
                        "star": star_list,
                        "gender": gender,
                        "age": [np.nan]*len(star_list),
                        "date": date_list
                        }

            page_df = pd.DataFrame(column_dict)
            
            # dfにpage_dfを追加する
            df = pd.concat([df, page_df], axis=0, ignore_index=True)
            try:
                elem = driver.find_element(By.XPATH, '//*[@id="zv-top-reviews_307"]/div[7]/a[7]')
                elem.click()
            
            except:
                elem = driver.find_element(By.XPATH, '//*[@id="zv-top-reviews_307"]/div[7]/a[6]')
                elem.click()

        except:
            print("no more pages")
            break

    df.to_csv(f'../data/gunze/gunze_{product}.csv', index=False)
```
